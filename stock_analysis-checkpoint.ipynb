```python
import numpy as np
import pandas as pd


data=pd.read_csv("Downloads/stock data/stocks/A.csv")
data.head(10)
```

# Basic Exploration

```python
data.describe()

```

```python
data.isnull().sum()
data['Date']=pd.to_datetime(data['Date'])
data.head(10)
```

# NumPy Operations

```python
#daily returns
data['daily_returns']=(data['Close']-data['Open'])/data['Open']
#Mean of Daily Returns
mean_return = np.mean(data['daily_returns'])
print("Average Daily Return:", mean_return)
std_return=np.std(data['daily_returns'])
print('Daily Return Volatility', std_return)
```

```python
initial_investment = 10
daily_returns = data['daily_returns']

total_value = initial_investment * np.cumprod(1 + daily_returns)

data['Total_Value'] = total_value

print("Value on the last day:", data['Total_Value'].iloc[-1])
```

# Pandas Operations

```python
data['7d_rolling']=data['Close'].rolling(window=7).mean()
data['30d_rolling']=data['Close'].rolling(window=30).mean()
```

```python
data['Short_Term_Signal'] = np.where(
    data['Open'] > data['7d_rolling'],
    'Profit Expected',
    'Loss Expected'
)

max_volume = data['Volume'].max()
max_volume_day = data[data['Volume'] == max_volume].index

print("Highest trading volume:", max_volume)
print("Date(s) of highest volume:", max_volume_day)
```

# Visualization

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.plot(data.index, data['Close'], color='blue', label='Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Stock Close Price Over Time')
plt.legend()
plt.grid(True)

plt.figure(figsize=(5,5))
plt.plot(data.index, data['Close'], color='green', label='open Price')
plt.xlabel('Date')
plt.ylabel('Open  Price')
plt.title('Stock Open Price Over Time')
plt.legend()
plt.grid(True)
plt.show()
plt.show()

```

```python


plt.figure(figsize=(12,6))
plt.plot(data.index, data['Close'], color='blue', label='Close Price')
plt.plot(data.index, data['7d_rolling'], color='orange', label='7-Day Rolling Avg')
plt.plot(data.index, data['30d_rolling'], color='green', label='30-Day Rolling Avg')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Close Price vs 7-Day & 30-Day Rolling Averages')
plt.legend()
plt.grid(True)
plt.show()
```
